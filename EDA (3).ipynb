```python
df=df.loc[~df.duplicated(subset=['coaster_name','Location','Opening date'])] .reset_index(drop=True).copy()
```

Feature Understanding
    * Plotting Feature Distribution
       * Histogram
       * KDE 
       * Boxplot

```python
ax=df['Opening date'].value_counts().head(10)\
.plot(kind='bar',title='Top years Coasters Introduced')
ax.set_xlabel('Year_Introduced')
ax.set_ylabel('Count')
```

```python
# Extract the first number from the string (e.g., '6' from '6 mph...')
df['Speed'] = df['Speed'].str.extract(r'(\d+)')  # This gives strings or NaN

# Convert to numeric, coercing any invalid ones to NaN
df['Speed'] = pd.to_numeric(df['Speed'], errors='coerce')

# Optionally fill NaNs (you can choose 0 or leave them)
df['Speed'] = df['Speed'].fillna(0).astype(int)
```

```python
bx=df['Speed'].plot(kind='hist',color='blue',bins=20,title='Coaster Speed(mph)')
bx.set_xlabel('Speed(mph)')
```

```python
bx=df['Speed'].plot(kind='kde',color='red',title='Coaster Speed(mph)')
bx.set_xlabel('Speed(mph)')
```

```python
df.head()
```

```python
df.plot(kind='scatter',x='Speed',y='height_ft',title='Coaster Speed vs Height')
plt.show()
```

```python
sns.scatterplot(x='Speed',y='height_ft',data=df)
```

```python
k=sns.pairplot(df,vars=['Opening date','Speed','height_ft','Inversions_clean','Gforce_clean'])
k._legend.remove()
plt.show()
```

```python
# Convert 'Opening date' to datetime if it's not already
df['Opening date'] = pd.to_datetime(df['Opening date'], errors='coerce')
g=sns.pairplot(df, vars=['Opening date', 'Speed', 'height_ft', 'Inversions_clean', 'Gforce_clean'], hue='Type')
g._legend.remove()
plt.show()
```

```python
df_corr=df[['Opening date','Speed','height_ft','Inversions_clean','Gforce_clean']].dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr,annot=True)
```
